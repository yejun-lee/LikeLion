````python
fig = plt.figure(figsize=(12, 12), facecolor='white')

# t1 = np.linspace(0, 2*np.pi, 100)
t1 = np.arange(0.0, 2 * np.pi, 0.1)

axis1 = fig.add_subplot(2, 2, 1)
axis1.plot(t1, np.sin(t1), 'g', label='$y=sin(t+2\pi)$')
axis1.legend()

axis2 = fig.add_subplot(2, 2, 2)
axis2.plot(t1, np.cos(t1), 'r')

axis3 = fig.add_subplot(2, 2, 3)
axis3.plot(t1, np.sinh(t1), 'b')

axis4 = fig.add_subplot(2, 2, 4)
axis4.plot(t1, np.cosh(t1), 'k')
````

zip(*반복가능한 자료형 여러 개 입력가능)

(예제!)
```
>>> list(zip([1, 2, 3], [4, 5, 6]))
[(1, 4), (2, 5), (3, 6)]
>>> list(zip([1, 2, 3], [4, 5, 6], [7, 8, 9]))
[(1, 4, 7), (2, 5, 8), (3, 6, 9)]
>>> list(zip("abc", "def"))
[('a', 'd'), ('b', 'e'), ('c', 'f')]
```

````python
team_score = { "toto":[1500,3000,5000,7000,5500],
              "apple":[4000,5000,6000,5500,4500],
              "gildong":[2000,2500,3000,4000,3000],
              "catanddog":[7000,5000,3000,5000,4000]}

team_df = pd.DataFrame(team_score,
                   columns=['catanddog', 'toto', 'apple', 'gildong'], 
                      index=idx)

team_df
````

````python
team_df[team_df.index > '2021-06-23']
````

````python
# 열 인덱스
team_df.columns
````

````python
# 행 인덱스
team_df.index
````

````python
team_df.loc[:, 'toto':'apple']
````

````python
team_df.loc['2021-06-21', 'toto':'apple']
````

````python
team_df.loc['2021-06-21':'2021-06-23', 'toto':'apple']
````

````python
team_df[['toto', 'gildong']]
````

````python
## 열 기준 연산
print(team_df.sum() )
print("----")
print(team_df.mean() ) 
print("----")
````

````python
## 행 기준 연산
print(team_df.sum(axis=1))
print("----")
print(team_df.mean(axis=1))
print("----")
````

````python
# 날짜별 누적 합계
print(team_df.cumsum())
print(team_df.cumsum(axis=1))
````

#### 실습 1
1. DataFrame을 하나 만들어보자. (5행 열 2,3개)
2. DataFrame을 동일한 데이터로 만들며 index를 날짜로 넣어보자.
3. 2~3행을 선택해보자.
4. 4번째 행을 선택해보자.
5. 2행 2열부터 3행 3열까지 선택해보자.

````python
## 실습 1-1~2
name = ['솔고바이오', '한세엠케이', '네오오토', '엄지하우스']
price = ['993', '9490', '9990' ,'4885']
amount = ['4657527', '19598600', '3157509', '7231']

idx = pd.date_range(start='2021-06-21', end='2021-06-24')

Data_1 = {'종목명':name, '현재가':price, '거래량':amount}
Data_1 = pd.DataFrame(Data_1, index=idx)

Data_1
````

````python
## 실습 1-3
print(Data_1.iloc[1:3])
print('------')
print(Data_1.iloc[-1])
````

````python
## 실습 1-4
print(Data_1.loc['2021-06-22':'2021-06-23'])
print('------')
print(Data_1.loc['2021-06-24'])
````

````python
## 실습 1-5
print(Data_1.loc['2021-06-22':'2021-06-23', '현재가':'거래량'])
````

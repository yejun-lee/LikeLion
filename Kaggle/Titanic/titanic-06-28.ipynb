### 결측치 처리

```python
train['Sex'] = train['Sex'].map({'male':0, 'female':1}).astype(int)
test['Sex'] = test['Sex'].map({'male':0, 'female':1}).astype(int)
```

```python
## fillna() : 괄호 안의 값으로 결측치를 채워주겠다.
# Age_mean = random.choice(range(16, 55))
train['Age'] = train['Age'].fillna(21.7)

# Age_mean = random.choice(range(16, 55))
test['Age'] = test['Age'].fillna(41.3)
```

```python
train.isnull().sum()
```

```python
train['Embarked'] = train['Embarked'].fillna('Q')
test['Embarked'] = test['Embarked'].fillna('Q')

train.isnull().sum(), test.isnull().sum()
```

```python
train['Embarked'] = train['Embarked'].map({'S':0, 'C':1, 'Q':2}).astype(int)
test['Embarked'] = test['Embarked'].map({'S':0, 'C':1, 'Q':2}).astype(int)
```

```python
train['Embarked'].unique()
```

```python
sel = ['Pclass', 'SibSp', 'Parch', 'Embarked', 'Sex', 'Title']

# X : 2차 데이터, x : 1차 데이터

X_train = train[sel]
y_train = train['Survived']

X_test = test[sel]

X_train.head(), y_train.head(), X_test.head()
```

```python
### 01. 머신러닝 모델 객체 만들기
### 02. 머신러닝 모델 훈련시키기
### 03. 훈련시킨 모델로 예측하기
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()   # 01
tree.fit(X_train, y_train)        # 02
pred = tree.predict(X_test)       # 03
```

```python
### 케글에 제출
sub = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
sub['Survived'] = pred
sub.to_csv('first_sub.csv', index=False)
```

```python
train['Title'] = train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
test['Title'] = test.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train['Title'], train['Sex'])
```

```python
train['Title'] = train['Title'].replace(['Lady', 'Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

train['Title'] = train['Title'].replace('Mlle', 'Miss')
train['Title'] = train['Title'].replace('Ms', 'Miss')
train['Title'] = train['Title'].replace('Mme', 'Mrs')


test['Title'] = test['Title'].replace(['Lady', 'Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

test['Title'] = test['Title'].replace('Mlle', 'Miss')
test['Title'] = test['Title'].replace('Ms', 'Miss')
test['Title'] = test['Title'].replace('Mme', 'Mrs')
```

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
all_df = [train, test]

for dataset in all_df:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

train.head()
```

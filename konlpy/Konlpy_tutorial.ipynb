```python
# 데이터가 너무 많으면 못 불러올 수도 있음 -> readline() -> 판다스 데이터프레임
doc_ko = open('15_TheExtreme_utf8.txt').read()   
doc_ko[1:1000]
```

```python
from konlpy.tag import Okt
t = Okt()

doc_nouns = t.nouns(doc_ko)
print(len(doc_nouns))
doc_nouns[1:50]
```

```python
# doc_nouns를 nltk의 텍스트로 만들겠다. nltk의 기능을 사용하겠다. (plot, tokken)
ko = nltk.Text(doc_nouns, name='분노의 질주')  
print(type(ko))
print(len(ko.tokens))
```

```python
ko.vocab().most_common(20)   # 가장 많이 사용된 단어 20개만 출력
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
plt.figure(figsize=(18,12))
ko.plot(50)
plt.show()
```

```python
### 한글에서는 따로 불용어 사전이 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = ['분노', '영화', '액션', '시리즈', '더', 
              '그', '이', '것', '또', '좀', 
              '돈', '것', '다음', '질주', '그냥', 
              '분노의질주', '말', '뭐', '애', '나', '듯', '편', '볼', '점', '중', '로']

new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="분노의 질주2")
plt.figure(figsize=(12,6))
new_ko.plot(50)
len(new_ko)
```

```python
from konlpy.tag import Okt

t = Okt()

doc_dallas = open('달라스바이어스클럽_리뷰_06_17.txt').read()
nouns_dallas = t.nouns(doc_dallas)

f = open('stopwords_extreme.txt')

lines = f.readlines()

stop_words = []

for one in lines:
    stop_words.append(one.strip())

new_ko = []

for one_word in nouns_dallas:
    if one_word not in stop_words:
        new_ko.append(one_word)

new_ko = nltk.Text(new_ko, name='달라스바이어스클럽')

new_ko[:50]
```

```python
plt.figure(figsize=(15, 8))

new_ko.dispersion_plot(['스토리', '스케일', '레토'])
```

```python
Car_mask = np.array(Image.open("movie_image_white.jpg"))

data = new_ko.vocab().most_common(1000)

wc = WordCloud(background_color='white', 
               max_words=200,
               mask=Car_mask,   
               contour_width=3, 
               contour_color="steelblue", 
               font_path=path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
